# Data cleaning for tirestimmen dog sound tag info
This cleaning work includes sound tags from tierstimmen dog sounds only. Cat sound tags were generated by a different method- manual tagging, since it had a multiple data sources and had no reliable tag info (too many irrelevant or missing tags) so I had to mannually inspect them. 
### Cleaning tag info data scraped from ts_anno.py
- The goal is to extract sound taggs from tierstimmen info csv file.
- Pandas dataframe is created from raw csv files then cleaned. 
- Parsed descriptions and tags of each file written in german then extracts tags in english.
- Inspected sounds for certain tags then tags with similar sounds are consolidated into one tag.
- Tags categories are: barking, whimpering, howling, other(includes growling, cakling, etc). 
### Inspected sound files with certain tags then discarded bad files.
- Some sounds are not useful (such as licking and snoring) for identifying an animal, so sound files with these tags are discarded.
